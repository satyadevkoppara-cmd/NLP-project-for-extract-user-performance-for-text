```python
# ==========================================
# NLP USER PERFORMANCE ANALYZER (ROBUST)
# ==========================================

# Install libraries
!pip install nltk scikit-learn

import nltk
import re
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

# Download resources
nltk.download('stopwords')
nltk.download('vader_lexicon')

# Initialize tools
stop_words = set(stopwords.words('english'))
sia = SentimentIntensityAnalyzer()

# ------------------------------------------
# CLEAN TEXT FUNCTION (SAFE)
# ------------------------------------------
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()

    # remove stopwords
    filtered = [w for w in words if w not in stop_words]

    # if all words removed → fallback
    if not filtered:
        return text.strip()

    return " ".join(filtered)

# ------------------------------------------
# SENTIMENT ANALYSIS
# ------------------------------------------
def get_sentiment(text):
    score = sia.polarity_scores(text)['compound']

    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"

# ------------------------------------------
# PERFORMANCE MAPPING
# ------------------------------------------
def map_performance(sentiment):
    if sentiment == "Positive":
        return "High"
    elif sentiment == "Neutral":
        return "Average"
    else:
        return "Low"

# ------------------------------------------
# KEYWORD EXTRACTION (ERROR-PROOF)
# ------------------------------------------
def extract_keywords(text):
    words = text.split()

    if not words:
        return ["No meaningful keywords"]

    return words[:5]   # simple + safe

# ------------------------------------------
# MAIN LOOP
# ------------------------------------------
print("\n🔍 NLP User Performance Analyzer")
print("Type 'exit' to stop\n")

while True:
    user_input = input("Enter text: ")

    if user_input.lower() == "exit":
        print("Program stopped.")
        break

    # Step 1: Clean
    cleaned = clean_text(user_input)

    # Step 2: Sentiment
    sentiment = get_sentiment(cleaned)

    # Step 3: Performance
    performance = map_performance(sentiment)

    # Step 4: Keywords
    keywords = extract_keywords(cleaned)

    # Output
    print("\n--- OUTPUT ---")
    print("Original Text:", user_input)
    print("Cleaned Text:", cleaned)
    print("Sentiment:", sentiment)
    print("Performance:", performance)
    print("Keywords:", keywords)
    print("---------------------------\n")
```
